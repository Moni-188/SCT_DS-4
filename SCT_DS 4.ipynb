```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
target = df['Accident_severity']
```

```python
df['Accident_severity'].value_counts()
```

```python
sns.countplot(x='Accident_severity',data=df)
plt.title("Accident Severity Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot (x=df['Weather_conditions'],hue = target)
plt.xticks(rotation = 45)
plt.title("Weather Condition vs Accident severity")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Road_surface_type',data=df)
plt.xticks(rotation = 45)
plt.title("Accidents by Road Surface Condition")
plt.show()
```

```python
sns.countplot(x='Light_conditions',data=df)
plt.xticks(rotation = 45)
plt.title("Accidents by Light Condition")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x = "Types_of_Junction",data=df)
plt.xticks(rotation = 45)
plt.title("Accidents by Types of Junction")
plt.show()
```
